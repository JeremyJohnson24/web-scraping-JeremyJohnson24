```python
def urlExtraction(URL):
    current_response = requests.get(URL)
    current_soup = BS(current_response.text)
    return(current_soup)
```

```python
for eachURL in list_of_sites:
    current_extraction = urlExtraction(eachURL)
    print(current_extraction.findAll('p')[1])
```
